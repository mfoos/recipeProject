```python
newcol = []
for i in range(len(data.ingredsTrimmedNoNum)):
    temp = []
    idx = 0
    words = data.ingredsTrimmedNoNum[i].split()
    while idx < (len(words) - 1):
        temp.append(words[idx:idx + 2])
        idx += 1
    newcol.append(temp)
data['ngrams_n2'] = pd.Series(newcol, index = data.index)
# Could probs do without outside loop, because same thing happens to each row, dur
```

```python
all_ngrams = map(lambda x: ' '.join(x), all_ngrams)
```

```python
ngrams_df = pd.DataFrame({'counts': counts, 'ngrams': unique_ngrams})
```

```python
ngrams_df_clean = ngrams_df.replace(to_replace="\s*(Tbsp|tsp|oz[s|\s+]|lb|cups*|&|\s*-\s*|and|ground|fresh|bunch|can|diced|dried|small|medium|med\s|large|,)\s*", 
                                    value="", 
                                    regex=True).sort_values('counts')
```

```python
ngrams_df_clean.head()
```

```python
ngrams_df_clean = ngrams_df_clean[map(lambda x: pd.notnull(x), [re.search(" ",x) for x in ngrams_df_clean['ngrams']])]
ngrams_df_clean.head()
```

```python
word_list = list(itertools.chain(*map(lambda x: x.split(" "), data.ingredsTrimmedNoNum)))
words, counts = np.unique(word_list, return_counts=True)
pd.DataFrame({'counts': counts, 'word': words}).sort_values('counts').tail(15)
```

```python
single = list(words)

import re
single = set(map(lambda x: re.sub('\s*(Tbsp|tsp|oz$|^as$|lb|cups*|&|^-$|and|bunch|powder[ed]*|can|diced|dried|inch[es]*|small|medium|med$|large|,)',"",x), single))
double = set(ngrams_df_clean['ngrams'])
double.issuperset(single) # womp womp, 2-grams doesn't include all words as perfect-match
single.difference(double) # only below are relevant!
```

```python
glossary = list(single.union(double))
glossary.sort()
glossary = pd.DataFrame({"ingredient":glossary})
glossary['include'] = "YES"
# if assign by longest match, (thank you, Peter K), some cleaning can be skipped, but 'egg', 'tomatoes' etc still trouble
# maybe using multiple match return could work for 3-grams
glossary.head()
```

```python
final = list(glossary[glossary['include'] == "YES"]['ingredient'])
len(final)
```
