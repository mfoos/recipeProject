```python
import pandas as pd
```

```python
data = pd.read_csv('preprocessed_BBwebscraping.csv', header = 0)
#preprocess = pd.read_csv('preprocessed_BBwebscraping.csv', header = 0)
```

```python
newcol = []
for i in range(len(data.ingredsTrimmedNoNum)):
    temp = []
    idx = 0
    words = data.ingredsTrimmedNoNum[i].split()
    while idx < (len(words) - 1):
        temp.append(words[idx:idx + 2])
        idx += 1
    newcol.append(temp)
data['ngrams_n2'] = pd.Series(newcol, index = data.index)
# Could probs do without outside loop, because same thing happens to each row, dur
```

```python
data.head()
```

```python
ngram2_pool = []
for i in newcol:
    for j in i:
        ngram2_pool.append(j) # only identifies 2mers, duh
```

```python
ngrams = data['ngrams_n2']
```

```python
import itertools
import numpy as np
```

```python
all_ngrams = list(itertools.chain.from_iterable(ngrams))
```

```python
' '.join(all_ngrams[0])
```

```python
all_ngrams = map(lambda x: ' '.join(x), all_ngrams)
```

```python
unique_ngrams, counts = np.unique(all_ngrams, return_counts=True)
```

```python
ngrams_df = pd.DataFrame({'counts': counts, 'ngrams': unique_ngrams})
```
